```python
def CalculateEucledianDistance(X_train, X_test):
    euc = np.sqrt(np.sum(np.square(np.subtract(X_train, X_test)), axis=1))
#     euc = np.sort(euc)
    
    return euc
```

```python
def predict(X_train, Y_train, X_test, k):
    
    eucdist = CalculateEucledianDistance(X_train, X_test)
    sorteucdist = np.sort(eucdist)
    
    indexes = np.arange(len(Y_train))
    dictionary = {k : v for k, v in zip(eucdist, indexes)}
    
    karray = sorteucdist[:k]
    
#     minecudist = np.min(karray)
    classes = []
    for i in karray:
        index = dictionary[i]
        classes.append(Y_train[index])

#     predicted_class = Y_train[[index for index, dist in dictionary.items() if dist == mineucdist]]
    unique, counts = np.unique(classes, return_counts=True)
    counter = 0
    predicted_class = 0
    for i in range(len(unique)):
        if counts[i]>counter:
            predicted_class = unique[i]
        
    
    return predicted_class
    
```

```python
k=30
pc = predict(X_train, Y_train, X_test, k)
```
